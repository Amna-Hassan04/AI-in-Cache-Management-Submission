```python
from google.colab import files
from pathlib import Path
import os

# Upload files (CAAR.cc, EGAA.cc, T-MAP.cc, DAC.cc, ChronoEntropy.cc)
uploaded = files.upload()
policy_files = list(uploaded.keys())
print("Uploaded:", policy_files)

# Move uploaded files into ChampSim new_policies folder
BASE_DIR = Path("/content/cacheforge/cacheforge/ChampSim_CRC2")
POLICY_DIR = BASE_DIR / "new_policies"
for f in policy_files:
    os.replace(f, POLICY_DIR / f)
```

```python
INC_DIR = BASE_DIR / "inc"
LIB_PATH = BASE_DIR / "lib/config1.a"

compiled_bins = {}
for cc in policy_files:
    name = Path(cc).stem
    exe = BASE_DIR / f"{name}-config1"
    print(f"Compiling {name}...")
    !g++ -Wall -std=c++17 -I{INC_DIR} {POLICY_DIR/name}.cc {LIB_PATH} -o {exe}
    compiled_bins[name] = exe
```

```python
import re
import pandas as pd

TRACE_DIR = BASE_DIR / "traces"
workloads = [
    "astar_313B.trace.gz",
    "lbm_564B.trace.gz",
    "mcf_250B.trace.gz",
    "milc_409B.trace.gz",
    "omnetpp_17B.trace.gz",
]

results = []
for name, exe in compiled_bins.items():
    for trace in workloads:
        trace_path = TRACE_DIR / trace
        print(f"Running {name} on {trace}...")
        out = os.popen(f"{exe} -warmup_instructions 1000000 -simulation_instructions 10000000 -traces {trace_path}").read()

        # Extract cumulative IPC
        m = re.search(r"CPU 0 cumulative IPC: ([0-9.]+)", out)
        ipc = float(m.group(1)) if m else None
        results.append([name, trace, ipc])

df = pd.DataFrame(results, columns=["Policy", "Trace", "IPC"])

```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Path to where your log_*.txt files are saved
LOG_DIR = Path("/content/cacheforge/cacheforge/ChampSim_CRC2")  # adjust if needed

results = []

for log_file in LOG_DIR.glob("log_*.txt"):
    text = log_file.read_text()

    # Extract policy and trace name from filename
    # Example filename: log_001_chronoentropy_lbm_564B.txt
    parts = log_file.stem.split("_")
    policy = "_".join(parts[1:-1])  # remove 'log' and trace stem
    trace = parts[-1]

    # Regex to get final cumulative IPC
    m = re.search(r"CPU 0 cummulative IPC[: ]+([0-9.]+)", text)
    if m:
        ipc = float(m.group(1))
        results.append([policy, trace, ipc])

df = pd.DataFrame(results, columns=["Policy", "Trace", "IPC"])
display(df)

# Average IPC per policy
avg_df = df.groupby("Policy")["IPC"].mean().reset_index().sort_values("IPC", ascending=False)
print("=== Average IPC per Policy ===")
display(avg_df)

# Plot: IPC by policy + trace
plt.figure(figsize=(12,6))
for policy, group in df.groupby("Policy"):
    plt.plot(group["Trace"], group["IPC"], marker="o", label=policy)

plt.xlabel("Trace")
plt.ylabel("IPC")
plt.title("Policy Performance per Trace")
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Plot: Average IPC bar chart
avg_df.plot(x="Policy", y="IPC", kind="bar", figsize=(10,5), legend=False)
plt.ylabel("Average IPC")
plt.title("Average IPC per Policy")
plt.show()
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

log_dir = Path("/content/cacheforge/cacheforge/ChampSim_CRC2")

def parse_log(file):
    text = Path(file).read_text(errors="ignore")
    # IPC
    ipc_match = re.search(r"cummulative IPC: ([0-9.]+)", text)
    ipc = float(ipc_match.group(1)) if ipc_match else None
    # MPKI (if available)
    mpki = None
    llc_miss = re.search(r"LLC TOTAL\s+ACCESS:\s+(\d+)\s+HIT:\s+(\d+)", text)
    if llc_miss:
        access, hit = map(int, llc_miss.groups())
        miss = access - hit
        instr = int(re.search(r"instructions: (\d+)", text).group(1))
        mpki = (miss / instr) * 1000
    return ipc, mpki

rows = []
for file in log_dir.glob("*.txt"):
    ipc, mpki = parse_log(file)
    policy = file.stem
    # infer workload name from file or inside log
    wl_match = re.search(r"traces/(.*?).trace", Path(file).read_text(errors="ignore"))
    workload = wl_match.group(1) if wl_match else "unknown"
    rows.append({"Policy": policy, "Workload": workload, "IPC": ipc, "MPKI": mpki})

df = pd.DataFrame(rows)
print(df)

# Save clean table
df.to_csv("policy_vs_baseline.csv", index=False)
```

```python
best_policy = "001_chronoentropy"  # adjust if filename differs
baselines = ["lru", "ship", "ship++", "hawkeye_final", "red", "dancrc2"]

# Filter
plot_df = df[df["Policy"].str.contains(best_policy) | df["Policy"].isin(baselines)]

# IPC plot
plt.figure(figsize=(10,6))
for pol in plot_df["Policy"].unique():
    subset = plot_df[plot_df["Policy"] == pol]
    plt.plot(subset["Workload"], subset["IPC"], marker="o", label=pol)
plt.xlabel("Workload")
plt.ylabel("IPC")
plt.title("IPC Comparison: Best Policy vs Baselines")
plt.legend()
plt.grid(True)
plt.show()

# MPKI plot (if available)
if plot_df["MPKI"].notna().any():
    plt.figure(figsize=(10,6))
    for pol in plot_df["Policy"].unique():
        subset = plot_df[plot_df["Policy"] == pol]
        plt.plot(subset["Workload"], subset["MPKI"], marker="o", label=pol)
    plt.xlabel("Workload")
    plt.ylabel("MPKI")
    plt.title("MPKI Comparison: Best Policy vs Baselines")
    plt.legend()
    plt.grid(True)
    plt.show()
```
